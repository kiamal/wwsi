### Przygotowanie danych
BoxPlot przyjmuje dane wejściowe w postaci Array lub Vector. W tym celu tworzymy Pandas.Series dla każdego z wymiarów który chcemy wyświetlić.

```python
#recode price to float
listings['price'] = listings['price'].apply(lambda x: float(x.replace('$','').replace(',','')))

#prepare tuples for each room_type
frames = [(i,listings.loc[(listings['room_type']==i) & (listings['price']<1500)]['price']) for i in listings['room_type'].unique()]
name, price = [i[0] for i in frames], [i[1] for i in frames]
```

## BoxPlot - opis
Wykres pudełkowy (**boxplot**) przedstawia graficzną reprezentację miar pozycyjnych i dystrybucji zbioru danych. 

[Link do dokumentacji matplotlib.pyplot.boxplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html#matplotlib.pyplot.boxplot)

[Dodatkowe przykłady customizacji](https://matplotlib.org/stable/gallery/statistics/boxplot_demo.html#sphx-glr-gallery-statistics-boxplot-demo-py)

### BoxPlot - opis elemetów

- **Q1** - pierwszy kwartyl
- **Q3** - trzeci kwartyl
- **IQR** - odległość między kwartylem trzecim i pierwszym
- **Outliers** - punkty danych o wartościach większy niż 1.5 x IQR

## Box plot - przykłady
##### Poszczególne parametry i operacje opisane w komentarzach

```python
import matplotlib.pyplot as plt
%matplotlib inline

# zdefiniowanie obiektu i osi
fig, axs = plt.subplots(nrows=4, ncols=2, figsize=(30, 40), sharey=False)

# funkcja do nadawania nazw osi i nazwy wykresu
def set_labels(ax, title, vertical=True):
    
    #ustawienie etykiet osi X i Y w zależności od parametru 'vertical'
    ax.set_ylabel('price') if vertical else ax.set_xlabel('price')
    ax.set_xlabel('room type') if vertical else ax.set_ylabel('room_type')
    
    #ustawienie tytułu wykresu
    ax.set_title(title, size=26)

    
# Najprostszy wykres pudełkowy - przekazujemy jedynie zbiór danych
title = '1) Najprostszy wykres bez parametrów'
ax = axs[0,0]
ax.boxplot(price)
set_labels(ax, title)


# Wykres z dodanymi etykietami, zmienionym symbolem dla outlierów, przedstawiony w skali logarytmicznej
title = '2) Dodanie etykiet, skala logarytmiczna, zmiana symbolu outlierów'
ax = axs[0,1]
ax.boxplot(x=price, 
           labels=name, # dodanie etykiet na oś X
           sym='x') # zmiana symbolu outlierów
ax.set_yscale('log') #ustawienie skali osi Y na logarytmiczną
set_labels(ax, title)


# Wykres z wypełnionymi boxami, bez pokazywania outlierów
title = '3) Wypełnienie boxów, wyłączenie outlierów'
ax = axs[1,0]
ax.boxplot(x=price,
           patch_artist=True, # wypełnienie boxów
           showfliers=False,
           labels = name
          )
set_labels(ax, title)


# Wykres ze zmienioną szerokością boxów
title = '4) Zmiana szerokości boxów, zwężanie boxów przy medianie'
ax = axs[1,1]
ax.boxplot(x=price,
           patch_artist=True,
           showfliers=False,
           labels = name,
           widths = .9, #ustawienie szerokości boxów
           notch = True # zwężanie boxów przy medianie w zależności od przedziału ufności
          )
set_labels(ax, title)


# Wykres poziomy
title = '5) Wykres poziomy'
ax = axs[2,0]
ax.boxplot(x=price,
           patch_artist=True,
           showfliers=False,
           labels = name,
           widths = .9,
           vert=False # ustawienie orientacji poziomej
          )
set_labels(ax, title, False)


title = '6) Dodanie średnich, ukrycie daszków'
ax = axs[2,1]
ax.boxplot(x=price,
           patch_artist=False,
           showfliers=False,
           labels = name,
           widths = .6,
           vert=False,
           showmeans = True, # pokazanie średnich
           meanline=True, # pokazanie średnich jako linie
           showcaps = False # ukrycie daszków IQR 
          )
set_labels(ax, title, False)


title = '7) Styl boxów'
# definicja stylu boxa
boxprops = dict(linestyle='solid', # typ linii
                linewidth=.5, # szerokość
                facecolor='green') # kolor wypełnienia
ax = axs[3,0]
ax.boxplot(x=price,
           patch_artist=True,
           showfliers=False,
           labels = name,
           widths = .6,
           vert=False,
           boxprops=boxprops #przypisanie słownika stylu do właściwości boxa
          )
set_labels(ax, title, False)


title = '8) Styl pozostałych elementów'
# definicja stylu boxa
boxprops = dict(linestyle='solid', linewidth=.5, color='darkgoldenrod', facecolor='orange')
# definicja stylu linii mediany
medianprops = dict(linestyle='solid', linewidth=1, color='darkblue')
# definicja stylu linii wyznaczających IQR
whiskerprops = dict(linestyle='dotted', linewidth=.9, color='orange')
# definicja stylu daszkód IQR
capprops = dict(linestyle='dashed', linewidth=.9, color='orange')
# definicja stylu linii średniej
meanlineprops = dict(linestyle='dashed', linewidth=1, color='green')

ax = axs[3,1]
ax.boxplot(x=price,
           patch_artist=True,
           showfliers=False,
           labels = name,
           widths = .6,
           vert=False,
           meanline=True,
           showmeans=True,
           boxprops=boxprops, #przypisanie słowników stylów do poszczególnych elementów
           medianprops=medianprops,
           whiskerprops=whiskerprops,
           meanprops=meanlineprops,
           capprops=capprops
          )
set_labels(ax, title, False)

plt.show()
```

## Box plot - pełna lista konfigurowalnych parametrów

```python
# All Params

# definicja stylu boxa
boxprops = dict(linestyle='solid', linewidth=.5, color='darkgoldenrod', facecolor='orange', alpha=.7)
# definicja stylu linii mediany
medianprops = dict(linestyle='solid', linewidth=1, color='red')
# definicja stylu linii średniej
meanlineprops = dict(linestyle='dashed', linewidth=1, color='green')
# definicja stylu linii wyznaczających IQR
whiskerprops = dict(linestyle='dotted', linewidth=.9, color='orange')
# definicja stylu daszkód IQR
capprops = dict(linestyle='solid', linewidth=.9, color='orange')
# definicja stylu outlierów
flierprops = dict(marker='o', markerfacecolor='grey', markersize=3, markeredgecolor='none')

fig, ax = plt.subplots(figsize=(30,10))

ax.boxplot(
        # parameters
           x=price, # dane
           notch=False, # zawężanie przy medianie
           sym=None, # symbol outlierów
           vert=False, # orientacja pozioma / pionowa
           whis=2, # zmiana definicji rozpiętości IQR - domyślnie 1.5
           bootstrap=None, # bootstrapping
           usermedians=None, # podanie własnych wartości median
           conf_intervals=None, # zmiana przedziałów ufności dla zawężanych boxów
           positions=None, # ustawienie pozycji boxów
           widths=.6, # szerokość boxa
           patch_artist=True, #wypełnienie boxa
           labels = name, # etykiety
           manage_ticks=True, # automatyczne wyrównanie osi i etykiet
           autorange=False,    
           meanline=True, # rysowanie linii średniej
           zorder=None, # warstwa na której rysowany jest wykres
    
        # extra parameters
           showcaps=None, # pokazywanie daszków IQR
           showbox=None, # pokazywanie boxów
           showfliers=True, # pokazywanie oulierów
           showmeans=True, # pokazywanie średniej
           capprops=capprops, # właściwości daszków
           boxprops=boxprops, # właściwości boxów
           whiskerprops=whiskerprops, # właściowści linii IQR
           flierprops=flierprops, # właściwości outlierów
           medianprops=medianprops, # właściwości linii mediany
           meanprops=meanlineprops # właściwości linii średniej
          )
ax.set_ylabel('Typ pokoju', size=20)
ax.set_xlabel('Cena', size=20)
ax.set_title('Ceny wynajmu mieszkań na AirBnB', size=36)
ax.set_xlim((0,400))

plt.show()
```
